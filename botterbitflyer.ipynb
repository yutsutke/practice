現在のビットコインの価格を入手する

```python
import ccxt

exchange = ccxt.bitflyer()
symbol = 'BTC/JPY'

ticker = exchange.fetch_ticker(symbol)
current_price = ticker['last']
print(f'現在価格: {current_price} JPY')
```

ビットコインから５秒ごとにの価格を入手する

```python
import ccxt
import time
import matplotlib.pyplot as plt

exchange = ccxt.bitflyer()
symbol = 'BTC/JPY'
WAIT_TIME = 5 # 5秒待機
prices = []  # 空の価格リストを初期化

while True:
    try:
        time.sleep(WAIT_TIME) # 5秒待機

        ticker = exchange.fetch_ticker(symbol)
        current_price = ticker['last']
        print(f'現在価格: {current_price} JPY')
        prices.append(current_price)  # 価格をリストに追加

        plt.plot(prices)
        plt.xlabel('時間')
        plt.ylabel('価格 (JPY)')
        plt.title('ビットコイン/日本円の価格変化')
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")


```

*   現在の価格が直近１０個のデータの中で何番目にあたるかを計算する。
*   現在価格と直近10個の価格を小数点以下4桁まで丸める。浮動小数点数は、有限桁数で表現されるため、特定の数値に近似値として表現されることがあり誤差が生じることがあるため。



```python
import ccxt
import time
import matplotlib.pyplot as plt

exchange = ccxt.bitflyer()
symbol = 'BTC/JPY'
WAIT_TIME = 3 # 5秒待機
prices = []  # 空の価格リストを初期化

while True:
    try:
        time.sleep(WAIT_TIME) # 5秒待機

        ticker = exchange.fetch_ticker(symbol)
        current_price = ticker['last']
        print(f'現在価格: {current_price} JPY')
        prices.append(current_price)  # 価格をリストに追加

        # 直近10個の価格を取得し、現在価格が何番目にあたるかを計算
        if len(prices) > 10:
            last_ten_prices = prices[-10:]
            print(p for p in last_ten_prices)
            rank = sum(round(current_price, 4) <= round(p, 4) for p in last_ten_prices)        # 浮動小数点数は、有限桁数で表現されるため、特定の数値に近似値として表現されることがあります。これにより、非常に小さな数値の違いが数値の等値性を維持するために無視される場合があります。これが、たまに間違った計算結果が返される原因です。
            for p in ( p for p in last_ten_prices):
                print(p)
            print(f'直近10個の価格中で、現在の価格は{rank}番目に高いです。')

            plt.plot(prices)
            plt.xlabel('時間')
            plt.ylabel('価格 (JPY)')
            plt.title('ビットコイン/日本円の価格変化')
            plt.show()


 
    except Exception as e:
        print(f"An error occurred: {e}")
```
